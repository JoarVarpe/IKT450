```python
import glob
import os
import re
import shutil
import random
```

```python
fish_location = "PATH_OF_FISH"
all_files = os.listdir(fish_location)
all_files_str = " ".join(all_files)
m = re.findall(r"([^\s]+)", all_files_str)
n = re.findall(r"[0-9]+?(?=_)", all_files_str)
unique_id = set()
for id in n:
    unique_id.add(id)

list_of_groupings = []

for unique in unique_id:
    list_of_id = []
    for file in all_files:
        if file.startswith(str(unique) + "_"):
            list_of_id.append(file)
    list_of_groupings.append(list_of_id)
    
print(list_of_groupings)
```

```python
for grouping in list_of_groupings:
    for unique in unique_id:
        if len(grouping) % 2 == 1:
            print(grouping)
```

```python
#make each fish id a folder
#
fish_location = "PATH_OF_FISH"
fish_location_body = "PATH_OF_FISH_BODY"
fish_dir = os.path.join(fish_location, "body_id")
if not os.path.exists(fish_dir):
    os.mkdir(fish_dir)

random.seed(7)
split_ratio = 0.8
list_number = 0
random.shuffle(list_of_groupings)

split_dir = os.path.join(fish_dir, str(split_ratio).replace(".", "_"))
if not os.path.exists(split_dir):
        os.mkdir(split_dir)

train_dir = os.path.join(split_dir, "train")
if not os.path.exists(train_dir):
        os.mkdir(train_dir)
        
test_dir = os.path.join(split_dir, "test")
if not os.path.exists(test_dir):
        os.mkdir(test_dir)

        
        
for list in list_of_groupings:
    
    if list_number < len(list_of_groupings) * split_ratio:
        path = train_dir
        for i in range(len(list)):
            if "_r_" in list[i]:
                str_list = list[i].split("_r_")
                the_dir = str_list[0]
                the_path = os.path.join(path, the_dir)
                if not os.path.exists(the_path):
                    os.mkdir(the_path)

                original_fish_body = os.path.join(fish_location_body, list[i])
                new_fish_body = os.path.join(the_path, list[i])
                shutil.copy(original_fish_body, new_fish_body)
            else:
                str_list = list[i].split("_l_")
                the_dir = str_list[0]
                the_path = os.path.join(path, the_dir)
                if not os.path.exists(the_path):
                    os.mkdir(the_path)
                    
                original_fish_body = os.path.join(fish_location_body, list[i])
                new_fish_body = os.path.join(the_path, list[i])
                shutil.copy(original_fish_body, new_fish_body)
        
    else:
        path = test_dir
        for i in range(len(list)):
            if "_r_" in list[i]:
                str_list = list[i].split("_r_")
                the_dir = str_list[0]
                the_path = os.path.join(path, the_dir)
                if not os.path.exists(the_path):
                    os.mkdir(the_path)

                original_fish_body = os.path.join(fish_location_body, list[i])
                new_fish_body = os.path.join(the_path, list[i])
                shutil.copy(original_fish_body, new_fish_body)
            else:
                str_list = list[i].split("_l_")
                the_dir = str_list[0]
                the_path = os.path.join(path, the_dir)
                if not os.path.exists(the_path):
                    os.mkdir(the_path)
                    
                original_fish_body = os.path.join(fish_location_body, list[i])
                new_fish_body = os.path.join(the_path, list[i])
                shutil.copy(original_fish_body, new_fish_body)
                
    list_number += 1

```
